```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id="2e2a36d13bb84b66a5a8814e1b24f6ab",
    client_secret="2428851e4ec146a58c682ceb9592abe7"
))

# Search for the artist
result = sp.search(q=["The Weeknd","Brent Faiyaz"], type="artist", limit=1)

# Extract the artist ID

#this returns a json object so we need to subset it in order to get the data let's say artist id 

artist_id = result["artists"]["items"][0]["id"]
print(artist_id)
```

```python
artists = ["The Weeknd", "Brent Faiyaz"]
results = []

for name in artists:
    r = sp.search(q=name, type="artist", limit=1)
    results.append(r["artists"]["items"][0]["id"])

print(results)
```
